```python
import pandas as pd

# load dataset
df = pd.read_csv(r"C:\Users\Mohammed Kaif\OneDrive\Desktop\Retail_Demand_KPI_Dashboard\online_retail.csv")

# preview
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df[df['Quantity'] > 0]
```

```python
df = df.dropna(subset=['CustomerID'])
```

```python
df = df.drop_duplicates()
```

```python
remove_words = ['POSTAGE', 'DOTCOM POSTAGE', 'MANUAL', 'BANK CHARGES', 'CARRIAGE']

df = df[~df['Description'].isin(remove_words)]
```

```python
df['Revenue'] = df['Quantity'] * df['UnitPrice']
```

```python
df['Revenue'] = df['Revenue'].round(2)
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], dayfirst=True, errors='coerce')
```

```python
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['Day'] = df['InvoiceDate'].dt.day
```
